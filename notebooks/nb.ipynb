# Multi-Agent for Automated Optical Inspection 

### Agent Performance (Response Time + Accuracy) Plot

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample performance data
data = {
	"Agent": [
		"YOLOv11", 
		"Visual Agent", 
		"Manager Agent", 
		"Router Agent", 
		"System Agent", 
		"Research Agent", 
		"Reasoning Agent"
	],
	"Response Time (ms)": [5, 18, 20, 23, 30, 90, 120],
	"Accuracy (%)": [5, 28, 36, 45, 55, 80, 99.6],
}

df = pd.DataFrame(data)

# Create bar charts for response time and accuracy
fig, ax1 = plt.subplots(figsize=(8, 5))

# Bar chart for response time
ax1.bar(df["Agent"], df["Response Time (ms)"], color="#4A90E2")
ax1.set_ylabel("Response Time (ms)", color="#4A90E2")
ax1.set_title("Agent Performance (Response Time + Accuracy)")
ax1.tick_params(axis='y', labelcolor="#4A90E2")

# Rotate x-axis labels diagonally
ax1.set_xticklabels(df["Agent"], rotation=45, ha='right')

# Secondary axis for accuracy
ax2 = ax1.twinx()
ax2.plot(df["Agent"], df["Accuracy (%)"], color="#E94E77", marker='o', linewidth=2, label="Accuracy")
ax2.set_ylabel("Accuracy (%)", color="#E94E77")
ax2.tick_params(axis='y', labelcolor="#E94E77")

plt.tight_layout()
plt.grid(True)
plt.show()
```
